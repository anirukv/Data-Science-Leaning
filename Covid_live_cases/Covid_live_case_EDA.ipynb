```python
# visit mohfw website for dataset
# https://www.mohfw.gov.in/

import pandas as pd
import numpy as np
```

```python
df=pd.read_json('https://www.mohfw.gov.in/data/datanew.json')
df.tail()
```

```python
df=df.iloc[:-1,1:] #Last record is sum of all records. Not required
df.head()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
max(df['positive'])
```

```python
#total_states= np.arange(len(df['state_name']))
```

### Positive cases in each state 

```python
from matplotlib.pyplot import figure
figure(num=None, figsize=(23,19), dpi=70, facecolor='w', edgecolor='r')
plt.barh(df['state_name'],df['positive'],align='center', alpha=0.5,
         color=(1,0,0), edgecolor=(0.5,0.2,0.8))
plt.xlim(1,max(df['positive'])+10)
plt.xlabel('Positive number of Cases')
plt.ylabel('State')
plt.title('Positive Cases')
#plt.yticks(total_states, df['state_name'])
```

### Active Cases in each state 

```python
figure(num=None, figsize=(23,19), dpi=70, facecolor='w', edgecolor='r')
plt.barh(df['state_name'],df['active'],align='center', alpha=0.5,
         color=(0,1,0), edgecolor=(0.5,0.2,0.8))
plt.xlim(1,max(df['active'])+100)
plt.xlabel('Active number of Cases')
plt.ylabel('State')
plt.title('Active Cases')
```

### Total deaths wrt states

```python
figure(num=None, figsize=(23,19), dpi=70, facecolor='w', edgecolor='r')
plt.barh(df['state_name'],df['death'],align='center', alpha=0.5,
         color=(1,0,1), edgecolor=(0.58,0.16,0.76))
plt.xlim(1,max(df['death'])+10)
plt.xlabel('Total death')
plt.ylabel('State')
plt.title('Death')
```
