```python
numeric_list=['int64','float64'] #most are of this type
numerical_cols = list(df.select_dtypes(include=numeric_list).columns)
```

```python
df= df[numerical_cols]
```

```python
df=df.drop('Id',axis=1)
```

```python
df
```

```python
#It is a good practice to split train test to avoid overfitting

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(df.drop(labels=['SalePrice'],axis=1),df['SalePrice'],test_size=0.3,random_state=0)
```

```python
X_train
```

```python
X_train.isnull().sum()
```

```python
from sklearn.feature_selection import mutual_info_regression
#determine mutual info
mutual_info= mutual_info_regression(X_train.fillna(0), y_train)
mutual_info
```

```python
mutual_info=pd.Series(mutual_info)
mutual_info.index= X_train.columns
mutual_info.sort_values(ascending=False)
```

```python
mutual_info.sort_values(ascending=False).plot.bar(figsize=(15,5))
```

```python
from sklearn.feature_selection import SelectPercentile
selected_top=SelectPercentile(mutual_info_regression,percentile=20)
selected_top.fit(X_train.fillna(0), y_train)
```

```python
selected_top.get_support()
```
